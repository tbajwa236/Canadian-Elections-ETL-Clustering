## Tranformation of dataframes

```python
# Split the composite value 'election_id' into multiple values
# For example GE.43.2019 can be split into GE/43/2019, where
# GE = general election, 43 = 43rd election, 2019 = year of election

df_candidates[[
   'type_election', 'election_number', 'Year']] = df_candidates['election_id'].str.split('.', expand = True)

df_polldetails[[
    'type_election', 'election_number', 'Year']] = df_polldetails['election_id'].str.split('.',expand = True)

df_results[[
    'type_election', 'election_number', 'Year']] = df_results['election_id'].str.split('.',expand = True)

df_rejected_ballots[[
    'type_election', 'election_number', 'Year']] = df_rejected_ballots['election_id'].str.split('.',expand = True)
```

```python
# Rename columns with smaller/easier to read column names

df_candidates.rename(columns={'electoral_district_number_numéro_de_circonscription': 'electoral_district_number', 
                              'candidate_candidat': 'full_name', 
                              'candidate_first_name_prénom_du_candidat': 'first_name',
                              'candidate_middle_name_second_prénom_du_candidat': 'middle_name', 
                              'candidate_family_name_nom_de_famille_du_candidat': 'family_name',
                              'candidate_occupation_english_anglais': 'candidate_occupation',
                              'incumbent_indicator_indicateur_candidat_sortant': 'incumbent_indicator',
                              'elected_candidate_indicator_indicateur_du_candidat_élu': 'elected_candidate_indicator',
                              'candidate_province_english_anglais': 'candidate_province',
                              'party_abbreviation_english_anglais': 'candidate_party'
                             }, inplace=True)

df_polldetails.rename(columns={'electoral_district_number_numéro_de_circonscription': 'electoral_district_number', 
                               'polling_station_number_numéro_du_bureau_de_scrutin': 'poll_station', 
                               'polling_station_name_nom_du_bureau_de_scrutin': 'poll_station_name',
                               'electors_for_polling_station_électeurs_du_bureau': 'electors_per_station', 
                               'void_poll_indicator_indicateur_de_bureau_supprimé': 'void_poll',
                               'no_poll_held_indicator_indicateur_de_bureau_sans_scrutin': 'no_poll_held'
                              }, inplace = True)

df_results.rename(columns={'electoral_district_number_numéro_de_circonscription': 'electoral_district_number',
                           'polling_station_number_numéro_du_bureau_de_scrutin': 'poll_station',
                           'candidate_poll_votes_count_votes_du_candidat_pour_le_bureau': 'candidate_vote'
                          }, inplace = True)

df_ridings.rename(columns={'electoral_district_number_numéro_de_circonscription': 'electoral_district_number', 
                           'electoral_district_name_english_anglais': 'electoral_district_name'
                          }, inplace = True)

df_rejected_ballots.rename(columns={'electoral_district_number_numéro_de_circonscription': 'electoral_district_number', 
                                    'polling_station_number_numéro_du_bureau_de_scrutin': 'poll_station',
                                    'rejected_ballots_for_polling_station_bulletins_rejetés_du_bureau': 'rejected_ballots'
                                   }, inplace = True)
```

```python
# Retain the necessary columns from the 2015 and 2019 election

df_candidates = df_candidates[['Year', 'election_number', 
                               'event_number', 'electoral_district_number',
                               'ballot_sequence_number', 'first_name', 
                               'family_name', 'candidate_occupation', 
                               'incumbent_indicator','elected_candidate_indicator', 
                               'candidate_province', 'candidate_party']].query('Year>= "2015"')

df_polldetails = df_polldetails[['Year', 'event_number', 
                                 'electoral_district_number','poll_station', 
                                 'poll_station_name', 'electors_per_station',
                                 'void_poll', 'no_poll_held', 
                                 'merge_indicator']].query('Year >= "2015"')

df_results = df_results[['Year', 'event_number', 'electoral_district_number', 
                         'poll_station', 'ballot_sequence_number',
                         'candidate_vote']].query('Year >= "2015"')

df_rejected_ballots = df_rejected_ballots[[
                        'electoral_district_number', 'poll_station',
                        'rejected_ballots', 'Year']].query('Year >= "2015"')

# Only retain ridings from after 2013 boundary redistribution(most recent)
# as they apply to both, 2015 and 2019 elections
df_ridings = df_ridings[['redistribution_boundaries','electoral_district_number',
                         'electoral_district_name']].query('redistribution_boundaries == "2013"')
```

## Merging dataframes into a single/complete dataset

1. df_results.shape: (880919, 6) 
2. df_polldetails.shape: (150486, 9)
3. df_rejected_ballots.shape: (150486, 4)
4. df_candidates.shape: (3938, 12) 
5. df_ridings.shape: (338, 3)

#### Observation: 

Each electoral district(riding) has multiple polling stations. Each polling station has multiple candidates
which is represented by the ballot sequence number at the respective polling station. 

#### Merging strategy:
1. Merge df_polldetails and df_rejected ballots as they have the same number of rows. This will provide details 
   at the polling station level. 
   
2. Merge the dataframe created in step 1 with df_results to obtain details at a ballot sequence level. 
   This dataframe will not contain the details about the candidates or the ridings yet.

3. Merge df_candidates and df_ridings. 

4. Merge dataframes created in step 2 and 3 to obtain a complete dataset at the most granular level i.e. ballot level.


#### Step 1

```python
# Merge df_polldetails and df_rejected_ballots

polldetails_rejected = pd.merge(df_polldetails, 
                                df_rejected_ballots, 
                                how= 'left', 
                                left_on= ['Year', 'poll_station', 
                                          'electoral_district_number'], 
                                right_on = ['Year', 'poll_station', 
                                            'electoral_district_number'])
```

#### Step 2

```python
# Merging df_results and polldetails_rejected

results_polldetails_rejected = pd.merge(df_results, 
                                        polldetails_rejected, 
                                        how= 'left', 
                                        left_on= ['Year', 'event_number', 
                                                  'electoral_district_number', 'poll_station'], 
                                        right_on= ['Year', 'event_number', 
                                                   'electoral_district_number', 'poll_station'])
```

#### Step 3

```python
# Merge df_candidates and df_ridings

candidates_ridings = pd.merge(df_candidates, 
                              df_ridings, 
                              how='left', 
                              left_on=['electoral_district_number'], 
                              right_on = ['electoral_district_number'])
```

#### Step 4

```python
# Merge results_polldetails_rejected and candidates_ridings
# results_polldetails_rejected: (df_results, df_rejected_ballots and df_polldetails)
# candidates_ridings: (df_candidates and df_ridings)

df_complete = pd.merge(results_polldetails_rejected, 
                       candidates_ridings, 
                       how='left', 
                       left_on=['Year', 'event_number', 
                                'electoral_district_number', 'ballot_sequence_number'], 
                       right_on = ['Year', 'event_number', 
                                   'electoral_district_number', 'ballot_sequence_number'])
```
